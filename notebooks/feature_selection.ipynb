```python
# Import necessary libraries for data manipulation, visualization, and modeling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression, RFE
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

Feature Selection for Rent Median Prediction

```python
# Exclude identifier columns and target variable from features
exclude_cols = ['suburb', 'postcode', 'SA2', 'SA2 NAME', 'geometry']

# Define the target variable for rent median prediction
target_rent_median = '2022_rent_median'

# Add the target variable to the exclude list
exclude_cols += [target_rent_median]

# Define all potential features for rent median prediction
features_median = [col for col in rent_df.columns if col not in exclude_cols]

# Display the features list for verification
print("Features used for rent median prediction:")
print(features_median)
```

```python
# Perform EDA to understand the relationships between features and the rent median

# Calculate correlation matrix between features and the target variable
corr_matrix_median = rent_df[features_median + [target_rent_median]].corr()

# Extract correlations with the target variable
corr_with_target_median = corr_matrix_median[target_rent_median].drop(target_rent_median, errors='ignore')

# Visualize the correlations using a bar plot
plt.figure(figsize=(16, 8))
corr_with_target_median.sort_values(ascending=False).plot(kind='bar')
plt.title('Correlation of Features with Rent Median')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
# Use statistical tests to select features with the strongest relationships with rent median

# Separate features and target
X_median = rent_df[features_median]
y_median = rent_df[target_rent_median]

# Handle missing values
X_median = X_median.fillna(0)
y_median = y_median.fillna(y_median.mean())

# Standardize features
scaler_median = StandardScaler()
X_median_scaled = scaler_median.fit_transform(X_median)

# Select top 20 features using the f_regression statistical test
selector_median = SelectKBest(score_func=f_regression, k=20)
selector_median.fit(X_median_scaled, y_median)
selected_indices_median = selector_median.get_support(indices=True)
selected_features_univariate_median = X_median.columns[selected_indices_median]

print("Top features for rent median based on univariate selection:")
print(selected_features_univariate_median.tolist())
```

```python
# Use LassoCV to select features based on their coefficients for rent median prediction

# Initialize and fit the LassoCV model
lasso_median = LassoCV(cv=5, random_state=42)
lasso_median.fit(X_median_scaled, y_median)

# Get coefficients from the Lasso model
lasso_coefficients_median = pd.Series(lasso_median.coef_, index=X_median.columns)

# Select features with non-zero coefficients
selected_features_lasso_median = lasso_coefficients_median[lasso_coefficients_median != 0].index.tolist()

print("Features selected by Lasso Regression for rent median:")
print(selected_features_lasso_median)
```

```python
# Use Random Forest Regressor to determine feature importance for rent median prediction

# Initialize and fit the Random Forest model
rf_median = RandomForestRegressor(n_estimators=100, random_state=42)
rf_median.fit(X_median, y_median)

# Get feature importances from the Random Forest model
importances_median = pd.Series(rf_median.feature_importances_, index=X_median.columns)

# Select the top 20 features based on importance
selected_features_rf_median = importances_median.sort_values(ascending=False).head(20).index.tolist()

print("Top features for rent median based on Random Forest:")
print(selected_features_rf_median)
```

```python
# Use RFE with Random Forest to select features for rent median prediction

# Initialize the RFE model
rfe_selector_median = RFE(estimator=RandomForestRegressor(n_estimators=100, random_state=42), n_features_to_select=20, step=1)
rfe_selector_median.fit(X_median, y_median)

# Get the selected features
selected_features_rfe_median = X_median.columns[rfe_selector_median.support_].tolist()

print("Features selected by RFE for rent median:")
print(selected_features_rfe_median)
```

```python
# Combine features selected by different methods for rent median prediction

# Combine selected features into a set to avoid duplicates
selected_features_median = set(selected_features_univariate_median) | set(selected_features_lasso_median) | set(selected_features_rf_median) | set(selected_features_rfe_median)

# Convert the set back to a list
final_selected_features_median = list(selected_features_median)

print("Final selected features for rent median prediction:")
print(final_selected_features_median)
```

```python
# Create new features or transformations for rent median prediction

# For example, calculate rent median growth rates
for year in range(2001, 2023):
    current_year = f"{year}_rent_median"
    previous_year = f"{year - 1}_rent_median"
    growth_rate_col = f"{year}_rent_median_growth"
    if current_year in rent_df.columns and previous_year in rent_df.columns:
        rent_df[growth_rate_col] = (rent_df[current_year] - rent_df[previous_year]) / rent_df[previous_year]

# Add new growth rate features to the features list
growth_features_median = [col for col in rent_df.columns if '_rent_median_growth' in col]
final_selected_features_median += growth_features_median
final_selected_features_median = list(set(final_selected_features_median))
```

```python
# Prepare the final dataset for modeling rent median

# Extract the selected features and target variable
X_final_median = rent_df[final_selected_features_median]
y_final_median = rent_df[target_rent_median]

# Handle missing values
X_final_median = X_final_median.fillna(0)
y_final_median = y_final_median.fillna(y_final_median.mean())
```

Feature Selection for Rent Count Prediction

```python
# Exclude identifier columns and target variable from features
exclude_cols = ['suburb', 'postcode', 'SA2', 'SA2 NAME', 'geometry']

# Define the target variable for rent count prediction
target_rent_count = '2022_rent_count'

# Add the target variable to the exclude list
exclude_cols += [target_rent_count]

# Define all potential features for rent count prediction
features_count = [col for col in rent_df.columns if col not in exclude_cols and not col.startswith(('2024', '2025', '2026'))]

# Display the features list for verification
print("Features used for rent count prediction:")
print(features_count)
```

```python
# Perform EDA to understand the relationships between features and the rent count

# Calculate correlation matrix between features and the target variable
corr_matrix_count = rent_df[features_count + [target_rent_count]].corr()

# Extract correlations with the target variable
corr_with_target_count = corr_matrix_count[target_rent_count].drop(target_rent_count, errors='ignore')

# Visualize the correlations using a bar plot
plt.figure(figsize=(16, 8))
corr_with_target_count.sort_values(ascending=False).plot(kind='bar')
plt.title('Correlation of Features with Rent Count')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
# Use statistical tests to select features with the strongest relationships with rent count

# Separate features and target
X_count = rent_df[features_count]
y_count = rent_df[target_rent_count]

# Handle missing values
X_count = X_count.fillna(0)
y_count = y_count.fillna(y_count.mean())

# Standardize features
scaler_count = StandardScaler()
X_count_scaled = scaler_count.fit_transform(X_count)

# Select top 20 features using the f_regression statistical test
selector_count = SelectKBest(score_func=f_regression, k=20)
selector_count.fit(X_count_scaled, y_count)
selected_indices_count = selector_count.get_support(indices=True)
selected_features_univariate_count = X_count.columns[selected_indices_count]

print("Top features for rent count based on univariate selection:")
print(selected_features_univariate_count.tolist())
```

```python
# Use LassoCV to select features based on their coefficients for rent count prediction

# Initialize and fit the LassoCV model
lasso_count = LassoCV(cv=5, random_state=42)
lasso_count.fit(X_count_scaled, y_count)

# Get coefficients from the Lasso model
lasso_coefficients_count = pd.Series(lasso_count.coef_, index=X_count.columns)

# Select features with non-zero coefficients
selected_features_lasso_count = lasso_coefficients_count[lasso_coefficients_count != 0].index.tolist()

print("Features selected by Lasso Regression for rent count:")
print(selected_features_lasso_count)
```

```python
# Use Random Forest Regressor to determine feature importance for rent count prediction

# Initialize and fit the Random Forest model
rf_count = RandomForestRegressor(n_estimators=100, random_state=42)
rf_count.fit(X_count, y_count)

# Get feature importances from the Random Forest model
importances_count = pd.Series(rf_count.feature_importances_, index=X_count.columns)

# Select the top 20 features based on importance
selected_features_rf_count = importances_count.sort_values(ascending=False).head(20).index.tolist()

print("Top features for rent count based on Random Forest:")
print(selected_features_rf_count)
```

```python
# Use RFE with Random Forest to select features for rent count prediction

# Initialize the RFE model
rfe_selector_count = RFE(estimator=RandomForestRegressor(n_estimators=100, random_state=42), n_features_to_select=20, step=1)
rfe_selector_count.fit(X_count, y_count)

# Get the selected features
selected_features_rfe_count = X_count.columns[rfe_selector_count.support_].tolist()

print("Features selected by RFE for rent count:")
print(selected_features_rfe_count)
```

```python
# Combine features selected by different methods for rent count prediction

# Combine selected features into a set to avoid duplicates
selected_features_count = set(selected_features_univariate_count) | set(selected_features_lasso_count) | set(selected_features_rf_count) | set(selected_features_rfe_count)

# Convert the set back to a list
final_selected_features_count = list(selected_features_count)

print("Final selected features for rent count prediction:")
print(final_selected_features_count)
```

```python
# 8. Feature Engineering with Division by Zero Handling

import numpy as np

# Calculate rent count growth rates with zero division handling
for year in range(2001, 2023):
    current_year = f"{year}_rent_count"
    previous_year = f"{year - 1}_rent_count"
    growth_rate_col = f"{year}_rent_count_growth"
    if current_year in rent_df.columns and previous_year in rent_df.columns:
        # Replace zeros in the previous year with NaN to avoid division by zero
        denominator = rent_df[previous_year].replace(0, np.nan)
        # Calculate growth rate
        rent_df[growth_rate_col] = (rent_df[current_year] - rent_df[previous_year]) / denominator

# Handle infinities and NaNs in the growth rate features
growth_features_count = [col for col in rent_df.columns if '_rent_count_growth' in col]

# Replace infinities with NaN
rent_df[growth_features_count] = rent_df[growth_features_count].replace([np.inf, -np.inf], np.nan)

# Fill NaN values with zero or another appropriate value (e.g., mean of the column)
rent_df[growth_features_count] = rent_df[growth_features_count].fillna(0)

# Add new growth rate features to the features list
final_selected_features_count += growth_features_count
final_selected_features_count = list(set(final_selected_features_count))
```

```python
# 9. Prepare Data for Modeling with Data Cleaning

# Extract the selected features and target variable
X_final_count = rent_df[final_selected_features_count]
y_final_count = rent_df[target_rent_count]

# Handle infinite values and NaNs in features
X_final_count = X_final_count.replace([np.inf, -np.inf], np.nan)
X_final_count = X_final_count.fillna(0)  # Or use appropriate imputation

# Ensure all feature values are finite numbers
if not np.all(np.isfinite(X_final_count.values)):
    print("Warning: X contains infinite or NaN values even after preprocessing.")

# Convert data types to float64 for higher precision
X_final_count = X_final_count.astype(np.float64)
y_final_count = y_final_count.astype(np.float64)

# Handle missing values in the target variable
y_final_count = y_final_count.fillna(y_final_count.mean())
```

Model evaluation

```python
# Model for Rent Median Prediction

# Split the data into training and testing sets
X_train_median, X_test_median, y_train_median, y_test_median = train_test_split(X_final_median, y_final_median, test_size=0.2, random_state=42)

# Initialize and train the Random Forest model
model_median = RandomForestRegressor(n_estimators=100, random_state=42)
model_median.fit(X_train_median, y_train_median)

# Predict on the test set
y_pred_median = model_median.predict(X_test_median)

# Evaluate the model performance
print("Model Performance for Rent Median Prediction:")
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test_median, y_pred_median))
print("Mean Squared Error (MSE):", mean_squared_error(y_test_median, y_pred_median))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test_median, y_pred_median)))
print("R^2 Score:", r2_score(y_test_median, y_pred_median))
```

<h2>Process 2023-2024 dataframe</h2>

```python
# process 2023-2024 dataframe
newrent_df = pd.read_csv('../data/curated/rent_cleaned.csv')

# transform price_per_week to numerical
newrent_df['price_per_week'] = pd.to_numeric(newrent_df['price_per_week'], errors='coerce')

# group by postcode, give median and count
df_group = newrent_df.groupby('postcode').agg(
    YearlyMedian=('price_per_week', 'median'),
    YearlyCount=('price_per_week', 'count')
).reset_index()

postcode_suburb = newrent_df[['postcode', 'suburb']].drop_duplicates()

# merge suburb name to grouped data 
final_data = pd.merge(df_group, postcode_suburb, on='postcode', how='left')

# Sort the results by suburb
final_data = final_data[['suburb', 'postcode', 'YearlyCount', 'YearlyMedian']]

print(final_data)
final_data.to_csv('../data/curated/2023-24_rent_cleaned_postcode.csv', index=False)
```

```python
# combined 2 dataset
df1 = rent_df
df2 = final_data

df2.rename(columns={
    'YearlyCount': '2023.5_rent_count',
    'YearlyMedian': '2023.5_rent_median'}, inplace=True)

comb_df = pd.merge(df1, df2, how='left', on=['suburb', 'postcode'])
comb_df.head(6)
```

```python
# use growth rate to count null median of 2023.5 data
comb_df['2023.5_rent_median'] = comb_df['2023.5_rent_median'].fillna(
    comb_df['2022_rent_median'] * (1 + comb_df['2022_rent_median_growth']))

# rent count value isn't sufficient in 2023.5 data
comb_df['2023.5_rent_count'] = comb_df['2022_rent_count'] * (1 + comb_df['2022_rent_count_growth'])

print(comb_df[['2022_rent_median', '2022_rent_median_growth', '2023.5_rent_median', '2022_rent_count','2022_rent_count_growth', '2023.5_rent_count']].head())

comb_df.to_csv('../data/curated/model_predict.csv', index=False)
```
